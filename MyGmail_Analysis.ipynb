```python
import mailbox
import pandas as pd
import csv
```

```python
path_data = 'data/gmail_data.mbox'
mbox = mailbox.mbox(path_data)
```

```python
# print labels
for i, key in enumerate(mbox[0].keys()):
    print(i+1, key)
```

```python
with open("data/mbox.csv", "w") as outfile:
    writer = csv.writer(outfile)
    for i, message in enumerate(mbox):
        if i%100 == 0:
            print(i, end=' ')
        writer.writerow([message['subject'], message['from'], message['to'],
                         message['date'], message['X-Gmail-Labels']])
        if i >= 999:
            break
```
